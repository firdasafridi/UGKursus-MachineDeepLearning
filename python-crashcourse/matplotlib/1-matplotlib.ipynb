# Matplotlib Object Oriented Method

Kita telah melihat cara membuat plot menggunakan matplotlib dengan menggunakan fungsi-fungsi. Kita akan melakukannya ulang, nemun dengan cara yang lebih formal, yakni menggunakan Matplotlib's Object Oriented API. Ini artinya kita akan melakukan instatiate object "figure" dan kemudian memanggil method-methodnya dari objek tersebut.

## Membuat figure dengan Plot Manual

```python
# Buat objek figure (canvas kosong)
fig = plt.figure()

# buat dan tetapkan axis untuk figure tersebut
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# plot nilai2 kita kedalam axis
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

Kodingnya terlihat lebih kompleks, tapi keuntungannya adalah kita bisa memiliki kontrol secara penuh untuk menaruh dimana letak plot kita, dan kita juga bisa dengan mudah menambahkan axis-axis lainnya kedalam figure.

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.6, 0.1, 0.5, 0.8]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

## Membuat figure dengan Plot Otomatis (subplot)

Obbjek plt.subplots() akan melakukan manajerial axis-axis kita secara otomatis. kita tidak perlu menetapkan range dan posisi axis kita, biarkan Objek plt.subplots() yang melakukannya.

```python
# Serupa dengan object plt.figure(), jika kita menggunakan plt.subplots(), kita mendapatkan tuple figure dan axis
fig, axes = plt.subplots()

# gunakan axes untuk plotting data-data kita
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

Kita juga bisa menspesifikasikan jumlah dari baris dan kolom ketika kita membuat objek subplot()

```python
# kavas kosong dengan 1x2 subplot
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# Jika penasaran axes itu isinya apa, coba kita tampilkan.
axes
```

Kita bisa lihat bahwa axes merupakan sebuah array. Itu tandanya kita bisa melakukan iterasi untuk seluruh axes yang ada.

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# tampilkan figure  
fig
```

Sebuah permasalahan yang umum ditemui menggunakan subplot atau figures adalah adanya overlapping antar axes. Kita bisa gunakan method **fig.tight_layout()** atau **plt.tight_layout()**. Method ini akan melakukan pengaturan posisi secara otomatis sehingga tidak ada overlapping lagi antar axes.

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig    
plt.tight_layout()
```

### Legends

Kita bisa menggunakan parameter **label="label text"** saat melakukan plotting, kemudian memanggil fungsi legend di akhir snippet code

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x squared")
ax.plot(x, x**3, label="x cube")
ax.legend()
```

Legend bisa ditaruh ditempat-tempat tertentu dalam grafik kita dengan menambakan parameter 'loc' dalam fungsi legend. 'loc' menerima numerical code dari 1-9, detail dari makna angka-angka tersebut bisa dilihat di [documentation page](http://matplotlib.org/users/legend_guide.html#legend-location) 

```python
ax.legend(loc=1) # pojok kanan atas
ax.legend(loc=2) # pojok kiri atas
ax.legend(loc=3) # pojok kiri bawah
ax.legend(loc=4) # pojok kanan bawah

# Yang biasa digunakan
ax.legend(loc=0) # biarkan matplotlib mencari posisi yang terbaik
fig
```

## Setting colors, linewidths, linetypes

Matplot lib memberikan kita banyak sekali opsi untuk melakukan kostumisasi warna, besar garis, tipe garis. Seluruh syntaxnya serupa dengan MATLAB.

### Kostumisasi Warna seperti MATLAB

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

### Kostumisasi Warna Menggunakan Keyword dan RGB hex code

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

### Kostumisasi Garis dan Style Marker

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='purple', linewidth=3, linestyle='--', marker='o', markersize=15, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
```

### Parameter Kostumisasi Secara Lengkap

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

## Plot range

Kita bisa melakukan konfigurasi rentang dari axis kita menggunakan method `set_ylim` dan `set_xlim` di objek axis. Atau, kita bisa gunakan `axis('tight')` untuk secara otomatis mendapatkan rentang yang "tightly fitted"

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
